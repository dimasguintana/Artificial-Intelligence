# **IMPORT LIBRARY**

```
import random
import math
import copy
```

# **GENERATE POPULASI & DECODE KROMOSOM**



```
# FUNGSI INI MENG-GENERATE POPULASI BERDASARKAN x (BANYAK INDIVIDU) DAN y (PANJANG GEN)

def populasi(x, y):
  return [[random.randint(0, 1) for j in range(y)] for i in range(x)]
```

```
# FUNGSI INI MENGEMBALIKAN x1 & x2 DARI SEBUAH INDIVIDU BERDASARKAN GENNYA
# BATAS BAWAH = -1 & BATAS ATAS = 2 UNTUK x1
# BATAS BAWAH = -1 & BATAS ATAS = 1 UNTUK x2

def phenotype(populasi, i):
  populasi_str = [int for int in populasi[i]]

  mid = len(populasi_str) / 2

  l1 = [x for x in populasi_str[0:int(mid)]]
  l2 = [x for x in populasi_str[int(mid):len(populasi_str)]]

  # mencari nilai x1
  sigma = 0
  g = 0
  r1min = -1
  r1max = 2

  for i in range(0, len(l1)):
    a = 2 ** -(i+1)
    sigma += a
    g += l1[i] * a

  x1 = r1min + (((r1max - r1min) / sigma) * g)

  # mencari nilai x2
  sigma = 0
  g = 0
  r2min = -1
  r2max = 1

  for i in range(0, len(l2)):
    a = 2 ** -(i+1)
    sigma += a
    g += l2[i] * a

  x2 = r2min + (((r2max - r2min) / sigma) * g)

  return x1, x2
```

```
# FUNGSI INI MENGEMBALIKAN NILAI DARI SEBUAH RUMUS

def h(x1, x2):
  return math.cos(x1) * math.sin(x2) - (x1 / ((x2 ** 2) + 1))
```

# **PENGHITUNGAN FITNESS**

```
# FUNGSI INI MENGEMBALIKAN NILAI FITNESS DARI SEBUAH INDIVIDU

def f(h):
  if(h >= 0):
    return 1 / (h+1)
  else:
    return -h
```

```
# FUNGSI INI MENG-APPEND NILAI FITNESS TIAP INDIVIDU DARI SUATU POPULASI

def lf(pop):
  listFitness = []
  for i in range(0, len(pop)):
    x1, x2 = phenotype(pop, i)
    listFitness.append(f(h(x1, x2)))

  return listFitness
```

# **PEMILIHAN ORANG TUA**

```
# FUNGSI INI MENDAPATKAN NILAI PI DARI TIAP2 INDIVIDU

def pi(listFitness):
  global p
  globals()['p'] = []

  for i in range(0, len(listFitness)):
    p.append(listFitness[i]/sum(listFitness))

  return p
```

```
# FUNGSI INI MENGEMBALIKAN ORANG TUA YANG TELAH DISELEKSI MENGGUNAKAN ROULETTE WHEEL

def RouletteWheelSelection(p):
  r = random.uniform(0, 1)
  idv = -1
  
  while(r > 0):
    idv += 1
    r -= p[idv]

  return idv
```

# **CROSSOVER (PINDAH SILANG)**

```
# FUNGSI INI MENGEMBALIKAN OFFSPRING DARI KEDUA ORANG TUA

def crossover(pop, parent1, parent2):
  probability = random.randint(60, 70) / 100
  Random = random.uniform(0, 1)
  randomPoint = random.randint(2, len(pop[parent1])-1)
  offspring = ''

  if(Random <= probability):
    # menghasilkan offspring1 dari parent1 & parent2
    offspring = ''.join(str(x) for x in pop[parent1][0:randomPoint])
    offspring += ''.join(str(x) for x in pop[parent2][randomPoint:len(pop[parent1])])
    offspring1 = [int(x) for x in offspring]

    offspring = ''
    # menghasilkan offspring2 dari parent1 & parent2
    offspring = ''.join(str(x) for x in pop[parent2][0:randomPoint])
    offspring += ''.join(str(x) for x in pop[parent1][randomPoint:len(pop[parent1])])
    offspring2 = [int(x) for x in offspring]

    return offspring1, offspring2

  else:
    return pop[parent1], pop[parent2]
```

# **MUTASI**

```
# FUNGSI INI MENGEMBALIKAN HASIL MUTASI APABILA NILAI Random-NYA KURANG DARI PROBABILITY

def mutation(offspring1, offspring2):
  probability = 1/100
  Random = random.uniform(0, 1)

  if(Random <= probability):
    mutasi1 = [1 if offspring1[i] == 0 else 0 for i in range(len(offspring1))]
    mutasi2 = [1 if offspring2[i] == 0 else 0 for i in range(len(offspring2))]
    return mutasi1, mutasi2
  else:
    return offspring1, offspring2
```

# **STEADY-STATE BY FITNESS**

```
#FUNGSI INI MENGEMBALIKAN ARRAY POPULASI YANG SUDAH DI REPLACE INDIVIDU

def steadyFit(pop, m1, m2, listFitness):
  sortFitness = copy.deepcopy(listFitness)
  sortFitness.sort()

  min1 = 0
  min2 = 0

  for i in range(len(listFitness)):
    if(sortFitness[0] == listFitness[i]):
      min1 = i
      break
  for i in range(len(listFitness)):
    if(sortFitness[1] == listFitness[i]):
      min2 = i
      break

  pop[min1] = copy.deepcopy(m1)
  pop[min2] = copy.deepcopy(m2)

  return pop
```

# **INDIVIDU TERBAIK**

```
# FUNGSI INI MENGEMBALIKAN INDIVIDU DENGAN FITNESS TERBESAR

def bestInd(listFitness):
  sortFitness = copy.deepcopy(listFitness)
  sortFitness.sort()

  max = 0

  for i in range(len(listFitness)):
    if(sortFitness[len(listFitness)-1] == listFitness[i]):
      max = i

  return max
```

# **MAIN PROGRAM**

```
individu = 35
gen = 10
jmlGenerasi = 500

pop = populasi(individu, gen)

for i in range(jmlGenerasi):
  populasiFitness = lf(pop)

  cek = copy.deepcopy(populasiFitness[0])
  counter = 0

  for j in range(1, len(populasiFitness)):
    if(cek == populasiFitness[j]):
      counter += 1
  if(counter == individu-1):
    break

  parent1 = RouletteWheelSelection(pi(populasiFitness))
  parent2 = RouletteWheelSelection(pi(populasiFitness))

  offspring1, offspring2 = crossover(pop, parent1, parent2)

  mutasi1, mutasi2 = mutation(offspring1, offspring2)

  pop = steadyFit(pop, mutasi1, mutasi2, populasiFitness)
  populasiFitness = lf(pop)
  # popFit = copy.deepcopy(populasiFitness)
  # popFit.sort()

  print('Generasi ke-',(i+1))
  print('Populasi :', pop)
  print('Fitness  :', populasiFitness,'\n')
  # print('Fitness Sort', popFit)

bestId = bestInd(populasiFitness)

print('\n---GENERASI TERBAIK---\n')
print('Populasi terbaik:', pop,'\n')
print('Individu terbaik:',pop[bestId],'\n')
x1, x2 = phenotype(pop, bestId)
print('x1:', x1, '\nx2:', x2,'\n')
print('h:', h(x1, x2))
print('Fitness:',f(h(x1, x2)))
```
