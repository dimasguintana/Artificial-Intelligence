# **LIBRARY**

```
import pandas as pd
from collections import Counter
import numpy as np
from sklearn.preprocessing import StandardScaler
```

# **READ CSV**

```
df= pd.read_csv('Diabetes.csv')
```

# **PRE PROCESSING DATA**

```
# Splitting Data
X = df.drop('Outcome', axis=1)
y = df['Outcome'].values

# Replace 0 (Missing Values) with NaN
X = X.replace(0,np.nan,regex=True)

# # Impute Missing Values with Mean
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
X = imp_mean.fit_transform(X)
```

```
def fit(a1, a2, a3, a4):
  if a4 == len(X):
    X_train = X[a1:a2]
    y_train = y[a1:a2]
    X_test = X[a3:a4]
    y_test = y[a3:a4]
  else:
    X_train = np.concatenate([X[a1:a3], X[a4:a2]])
    y_train = np.concatenate([y[a1:a3], y[a4:a2]])
    X_test = X[a3:a4]
    y_test = y[a3:a4]

  return X_train, y_train, X_test, y_test
```

# **DISTANCE CALCULATION**

```
def manhattan(x1, x2):
  return np.sum(np.abs(np.array(x1) - np.array(x2)))
```

# **CLASSIFICATION KNN**

```
def learn(X_test):
  predicted = [distance(x) for x in X_test]
  return np.array(predicted)

def distance(x):
  dist = [manhattan(x, x_train) for x_train in X_train]
  sort_k = np.argsort(dist)[:k]
  label_k = [y_train[i] for i in sort_k]
  return Counter(label_k).most_common(1)[0][0]
```

# **CHOOSE THE BEST K**

```
# Accuracy 5-Fold Cross-Validation with K = 10 to K = 19
k = 3
accuracy_result = []
for i in range(11):
  total_accuracy = 0
  for j in range(5):
    if j == 0:
      X_train, y_train, X_test, y_test = fit(0, 614, 614, len(X))
    elif j == 1:
      X_train, y_train, X_test, y_test = fit(0, len(X), 460, 614)
    elif j == 2:
      X_train, y_train, X_test, y_test = fit(0, len(X), 306, 461)
    elif j == 3:
      X_train, y_train, X_test, y_test = fit(0, len(X), 153, 307)
    elif j == 4:
      X_train, y_train, X_test, y_test = fit(154, len(X), 0, 154)

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    predictions = learn(X_test)
    total_accuracy += (np.sum(predictions == y_test) / len(y_test)) * 100
  print('Akurasi dengan K = ', k, ':', total_accuracy / 5)
  accuracy_result.append([k, (total_accuracy / 5)])
  k += 2

def sortAccuracy(elem):
  return elem[1]
list_accuracy = sorted(accuracy_result, key=sortAccuracy, reverse=True)

print('Dengan K =', list_accuracy[0][0], ', merupakan K yang memiliki akurasi tertinggi, yaitu:', list_accuracy[0][1])
```
